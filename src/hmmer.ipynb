```python
import pandas as pd

def parse_domtblout(file_path):
    data = []
    with open(file_path) as f:
        for line in f:
            if line.startswith('#') or not line.strip():
                continue
            parts = line.strip().split()
            target_name = parts[0]            # ID белка
            pfam_domain = parts[3]            # Имя домена PFAM (например Helicase_C)
            pfam_accession = parts[4]         # PFXXXXX.XX
            evalue = float(parts[6])          # Полный E-value
            score = float(parts[7])           # Скора
            data.append([target_name, pfam_domain, pfam_accession, evalue, score])

    df = pd.DataFrame(data, columns=['Protein_ID', 'Pfam_Domain', 'Pfam_Accession', 'Evalue', 'Score'])
    return df

domtblout_path = "fly.domtblout"
df = parse_domtblout(domtblout_path)
selected_pfams = [
   'PF05499.15',
   'PF00856.31',
   'PF02373.25',
   'PF00850.22',
  'PF00439.28',
   'PF00089.29',
   'PF00090.22',
   'PF00096.29',
   'PF00104.33',
  'PF00226.34',
  'PF00105.21',
 'PF00130.25',
  'PF00145.20',
  'PF00224.24',
   'PF00250.21',
   'PF00271.34'
]

filtered_df = df[df['Pfam_Accession'].isin(selected_pfams)]
result_table = filtered_df.groupby(['Pfam_Domain', 'Pfam_Accession'])['Protein_ID'].apply(list).reset_index()
result_table.to_csv("epigenetic_genes_by_family.csv", index=False)

print(result_table.head())
```
